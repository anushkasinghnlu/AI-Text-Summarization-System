```python
from transformers import pipeline
```

```python
summarizer = pipeline(
    "summarization",
    model="sshleifer/distilbart-cnn-12-6"
)

print("Model loaded successfully!")
```

```python
text = """
Artificial Intelligence is rapidly transforming the world.
It is used in healthcare, finance, education, and transportation.
AI systems analyze large volumes of data and help organizations
make better and faster decisions. Machine learning and deep
learning are important parts of AI that allow systems to learn
from data and improve automatically over time.
"""
```

```python
summary = summarizer(text, max_length=60, min_length=20, do_sample=False)

print("Original Text:\n")
print(text)

print("\nGenerated Summary:\n")
print(summary[0]['summary_text'])
```

```python
user_text = input("Enter a long paragraph to summarize:\n")

result = summarizer(user_text, max_length=60, min_length=20, do_sample=False)

print("\nSummary:\n")
print(result[0]['summary_text'])
```

```python
with open("summary_output.txt", "w") as file:
    file.write(result[0]['summary_text'])

print("Summary saved to file!")
```
